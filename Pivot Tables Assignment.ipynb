```python
import pandas as pd
```

```python
df = pd.read_csv("autos.csv")
```

# Printing out the first 8 lines

```python
df.head(8)
```

# Returning a data frame containing the minimum price of each car

```python
pd.pivot_table(df, index = "make", values = "price", aggfunc = "min")
```

# The smallest standard deviation of the prices

```python
pd.pivot_table(df, index = "make", values = "price", aggfunc = "std")
```

# Returning the last 2 rows of each group of car make and fuel_type 

```python
lr = df.groupby(['make','fuel_type'])
ltr = lr.apply(lambda x: x.tail(2))
ltr
```

# Using the split-apply-combine function

```python
volvo_data = df[df['make'] == 'volvo']
volkswagen_data = df[df['make'] == 'volkswagen']
```

```python
volvo_price_mean = volvo_data['price'].mean()
volkswagen_price_mean = volvo_data['price'].mean()
result = pd.DataFrame({
    'Car': ['volvo','volkswagen'],
    'Mean Price': [volvo_price_mean, volkswagen_price_mean]
})
result
```

# Separate Dictionaries and Joinng them together

```python
standard_diviation = df.groupby('make')['price'].agg(['std']).round(2)
minimum_place = df.groupby('make')['price'].agg(['min'])

df1 = pd.DataFrame(standard_diviation)
df2 = pd.DataFrame(minimum_place)
```

```python
mergin_df = pd.merge(df1, df2, on='make')
mergin_df
```
